3. Correlation matrix. (to just get an idea)

```python
fig, ax = plt.subplots(figsize=(10,5))
corrMatrix = dt.corr()
sns.heatmap(corrMatrix, annot=True)
plt.show()
```

__Observations__:
- Its bit werid that the nearest MRT distance has a negative impact on the price of house. One possible hypothesis could be the noise and disturbance of MRT maybe the reason of the low price of the house (maybe people don't want buy houses close to these activities and they only do it because its cheap because we have negative correlation there (-0.67))
- As expected the convinience of the nearby stores have positive impact on the price of house. 
- Interestingly longitude and latitude have an a positive impact on the house (maybe it is because of locality) 
- Its interesting to see that age of house has a negative impact on price (as the house gets older the price goes down) but the impact is very low.
